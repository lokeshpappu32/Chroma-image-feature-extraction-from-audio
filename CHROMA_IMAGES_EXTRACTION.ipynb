```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio signal
audio_path = '/content/drive/MyDrive/Dataset Child/Female/F1_01_01.wav'
signal, sr = librosa.load(audio_path)

# Extract chroma features
chroma_features = librosa.feature.chroma_stft(y=signal, sr=sr)

# Plot the chroma features as an image
plt.figure(figsize=(10, 4))
librosa.display.specshow(chroma_features, x_axis='time', y_axis='chroma', cmap='coolwarm')
plt.colorbar()
plt.title('Chroma Features')
plt.tight_layout()
plt.show()
```

```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Path to the folder containing audio files
folder_path = '/content/drive/MyDrive/Mozilla Dataset Adult/Female'

# Output folder to save the images
output_dir = '/content/drive/MyDrive/CHROMA_IMAGES_ADULT_FEMALE'
# Initialize a counter variable
image_count = 0

# Iterate over the audio files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.mp3'):
        # Load the audio signal
        audio_path = os.path.join(folder_path, filename)
        signal, sr = librosa.load(audio_path)

        # Extract chroma features
        chroma_features = librosa.feature.chroma_stft(y=signal, sr=sr)

        # Plot the chroma features as an image
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(chroma_features, x_axis='time', y_axis='chroma', cmap='coolwarm')
        plt.colorbar()
        plt.title('Chroma Features')
        plt.tight_layout()
         # Save the image to the output directory with a new name
        new_filename = f'Adult-Female_{image_count}.png'
        output_path = os.path.join(output_dir, new_filename)
        plt.savefig(output_path)
        plt.close()
        
        # Increment the image count
        image_count += 1
```
